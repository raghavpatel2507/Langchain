```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
```

```python
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain.schema import Document

# Create LangChain documents for IPL players

doc1 = Document(
        page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style and fitness, he has led the Royal Challengers Bangalore in multiple seasons.",
        metadata={"team": "Royal Challengers Bangalore"}
    )
doc2 = Document(
        page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm demeanor and ability to play big innings under pressure.",
        metadata={"team": "Mumbai Indians"}
    )
doc3 = Document(
        page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicketkeeping, and leadership are legendary.",
        metadata={"team": "Chennai Super Kings"}
    )
doc4 = Document(
        page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his yorkers and death-over expertise.",
        metadata={"team": "Mumbai Indians"}
    )
doc5 = Document(
        page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his quick fielding and match-winning performances make him a key player.",
        metadata={"team": "Chennai Super Kings"}
    )
```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    embedding_function=embedding,
    persist_directory='my_chroma_db',
    collection_name='sample'
)
```

```python
# add documents
vector_store.add_documents(docs)
```

```python
retriver=vector_store.as_retriever(search_kwargs={'k':2})
```

```python
result=retriver.invoke("who is rohit sharama")
print(result[0].page_content)
```

# 2.MMR RETRIVER

```python
retriver=vector_store.as_retriever(search_type='mmr',search_kwargs={'k':3,'lambda_mult':0.5})
#lambda_mult-relevance diversity-1 means normal similarity search,range:0-1
```

```python
result=retriver.invoke("who is rohit sharama")
```

```python
for i,doc in enumerate(result):
  print(f'\n---Result{i+1}---')
  print(doc.page_content)
```

# 3.Contextual Compression Retriever

```python
from langchain.retrievers.contextual_compression import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
```

```python
base_retriver=vector_store.as_retriever(search_kwargs={'k':2})
```

```python
from langchain_groq import ChatGroq
```

```python
from google.colab import userdata
GROQ_API_KEY=userdata.get('GROQ_API_KEY')
```

```python
llm=ChatGroq(model_name="llama-3.3-70b-versatile",api_key=GROQ_API_KEY)
compressor=LLMChainExtractor.from_llm(llm)
```

```python
compression_retriver=ContextualCompressionRetriever(
    base_retriever=base_retriver,base_compressor=compressor
)
```

```python
result=compression_retriver.invoke("what is virat kohli how to diffrent from others players")
```

```python
for i,doc in enumerate(result):
  print(f'\n---Result{i+1}---')
  print(doc.page_content)
```

# 4.Malti-Query Retriver

```python
# demo code
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain.chat_models import ChatOpenAI

# 1. Load your vector store (e.g., FAISS)
embedding_model = OpenAIEmbeddings()
vectorstore = FAISS.load_local("faiss_index", embeddings=embedding_model)

# 2. Set up your base retriever
retriever = vectorstore.as_retriever()

# 3. Use MultiQueryRetriever with LLM (e.g., GPT-4)
llm = ChatOpenAI(temperature=0, model_name="gpt-4")
multi_query_retriever = MultiQueryRetriever.from_llm(retriever=retriever, llm=llm)

# 4. Example query
query = "What are the symptoms of diabetes?"

# 5. Retrieve documents using multi-query retrieval
docs = multi_query_retriever.get_relevant_documents(query)

# 6. Print results
for doc in docs:
    print(doc.page_content)
```
